```python
import pandas as pd
```

```python
stocks = pd.read_csv('http://bit.ly/smallstocks', parse_dates = ['Date'])
```

```python
stocks
```

```python
stocks.index
```

```python
sgb = stocks.groupby(['Symbol', 'Date'])
```

```python
sgb.sum()
```

```python
sgb.describe()
```

```python
sgb.agg(['sum', 'max'])
```

```python
grpFrame = sgb.Close.agg(['min', 'max', 'mean'])
```

```python
grpFrame
```

```python
grpFrame.unstack()
```
